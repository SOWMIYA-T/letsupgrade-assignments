Plot using sns.relplot, line plot, box plot on fmri data set


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fmri=sns.load_dataset('fmri')
fmri.head()
```

Relplot

```python
sns.relplot(x='timepoint', y='signal', kind='line', data=fmri)
```

```python
sns.relplot(x='timepoint', y='signal',hue='event', kind='line', data=fmri)
```

```python
sns.relplot(x='timepoint', y='signal',hue='event',size='region', kind='line', data=fmri)
```

```python
sns.relplot(x='timepoint', y='signal',hue='region', size='event', kind='line', data=fmri, ci='sd')
```

```python
sns.relplot(x='timepoint', y='signal',hue='event', size='event', col='region', kind='line', data=fmri)
```

```python
sns.relplot(x='timepoint', y='signal',hue='event', size='event', col='region',height=5, aspect=.6, kind='line', data=fmri)
```

BoxPlot

```python
sns.catplot(x='region',y='signal',kind='box',data=fmri)
```

```python
sns.catplot(x='event',y='timepoint',kind='box',data=fmri,hue='region')
```

```python
sns.boxplot(x='timepoint',y='signal',hue='event',data=fmri)
```

LinePLot

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)
```

```python
sns.lineplot(x='timepoint',y='signal',hue='event',data=fmri)
```

```python
sns.lineplot(x='timepoint', y='signal',hue='event', size='event', data=fmri)
```
